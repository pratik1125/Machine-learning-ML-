```python
#Easiet way
apr_df=pd.read_csv('/content/puchased_data.csv',header= None)
```

```python
apr_df
```

```python
#The Apriori libray requires our dataset to be in form of a list of lists,
#where the whole dataset is a big list and each trascation in the dataset is an inner list within the outer big list.
```

```python
records=[]
for i in range(0,len(apr_df)):
  #for j in range(0,4):
   # records.append(str(apr_df.values[i,j]))
   records.append([str(apr_df.values[i,j]) for j in range(0,4)])
print(records)
```

```python
!pip install apyori
from apyori import apriori
```

```python
#Generate the rules
association_rules=apriori(records,min_support=0.5,min_confidence=0.75)
#convert rules into list
association_results=list(association_rules)
print(association_results)
```
