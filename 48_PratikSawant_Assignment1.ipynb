```python
df = pd.read_csv("Assignement_Data2 - Assignement_Data2.csv")
df.head()
```

```python
#Simple bar chart
plt.bar(df.index, df.col_1,edgecolor = 'k')
plt.xticks(df.index)
plt.show()
```

```python
#Horizontal bar chart
plt.barh(df.index, df.col_1,edgecolor = 'k')
plt.yticks(df.index)
plt.show()
```

```python
#Stacket Bar
plt.bar(df.index, df.col_1,edgecolor = 'k')
plt.bar(df.index, df.col_2, bottom = df.col_1,edgecolor = 'k')
plt.bar(df.index, df.col_3, bottom = df.col_1 + df.col_2,edgecolor = 'k')

plt.legend(labels = ['Col_1', 'Col_2', 'Col_3'])
plt.show()
```

```python
#Line Chart
plt.plot(df['col_1'])
plt.show()
```

```python
plt.plot(df['col_1'])
plt.plot(df['col_2'], ls = "--")

plt.legend(labels = ["Col_1", "Col_2"])
plt.show()
```

```python
plt.plot(df['col_3'], marker = "o")
plt.plot(df['col_4'], c = 'r', lw = '2', zorder = -1)

plt.legend(labels = ["Col_3", "Col4"])
plt.show()
```

```python
#Area plot
df.plot.area()
plt.show()
```
