```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

```python
#Subplots
plt.figure(figsize = (14, 10))

#line chart
plt.subplot(2,3,1)
data = df.groupby("Pclass")['Fare'].mean()
plt.plot(data.index, data, marker='o')
plt.xticks([1, 2, 3])
plt.xlabel('Class')
plt.ylabel('Fare')

#bar chart
plt.subplot(2,3,2)
data = df['Survived'].value_counts()
plt.bar(height = data, x = data.index, color = ['red', 'blue'])
plt.xticks([0, 1])

#Scatter chart
plt.subplot(2,3,3)
plt.scatter(df['Age'], df['Fare'], c = df['Pclass'], cmap = 'viridis')
plt.xlabel('Age')
plt.ylabel('Fare')

#Area
plt.subplot(2,3,4)
index = [1, 2, 3]
count = df.Pclass.value_counts().reset_index().sort_values(by = 'index')['Pclass']
total = df.groupby("Pclass")['Fare'].sum()
plt.stackplot(index, count, total)
plt.xticks(index)
plt.legend(['Count', 'Revenue Generated'])


#pie chart
plt.subplot(2,3,5)
data = df['Pclass'].value_counts()
plt.pie(data, autopct="%.2f")
plt.legend(labels = data.index)
plt.title("Class distribution")

#histogram
plt.subplot(2,3,6)
plt.hist(df['Age'])
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```
