```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("/content/result.csv")
df.info()
```

```python
df.head()
```

```python
x=df.drop("Result",axis=1)
```

```python
y=df.Result
```

```python
print(x)
print(y)
```

```python
KNN=KNeighborsClassifier(n_neighbors=2)
```

```python
KNN.fit(x,y)
```

```python
pred=KNN.predict([[6,8]])
pred
```

```python
KNN.predict([[2,3]])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, confusion_matrix,accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
KNN.fit(x_train,y_train)
```

```python
y_pred=KNN.predict(x_test)
```

```python
y_pred
```

```python
cmf=confusion_matrix(y_test,y_pred)
tm,fp,fn,tp=cmf.ravel()
```

```python
cmf
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
pre=precision_score(y_test,y_pred,average=None)
pre
```

```python
recall=recall_score(y_test,y_pred,average=None)
recall
```

```python
f1score=f1_score(y_test,y_pred,average=None)
f1score
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred,target_names=['0','1']))
```
