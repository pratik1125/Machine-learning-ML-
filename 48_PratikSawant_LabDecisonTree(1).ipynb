```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=4,random_state=42)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
#With random_state=0,we get the same train and test sets across different executions , with random_state=42, we get the same train and set across different executions,
#but in this time,the train sets are different from the previous case with random_state=0.
```

```python
df=pd.read_csv('/content/golf_df.csv')
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x=df.drop("Play",axis=1)
```

```python
y=df.Play
```

```python
onehot=pd.get_dummies(x)
```

```python
onehot
```

```python
from sklearn import tree
```

```python
clf=tree.DecisionTreeClassifier()
```

```python
clf_train=clf.fit(onehot,y)
```

```python
import matplotlib.pyplot as plt
```

```python
tree.plot_tree(clf_train)
```

```python
tree.plot_tree(clf_train,filled=True,rounded=True,feature_names=list(onehot.columns.values))
plt.savefig('out_playgolf.pdf')
plt.plot()
```

```python
##More Gini Index=More Impure =1(Impure Class)
##Less Gini Index = More pure Class
##Gini Index=(Pure class) 0
##i.e its a leaf node and futher can't be split

```

```python
prediction=clf_train.predict([[False,0,0,1,0,1,0,0,1]])
prediction
```

```python
x_train,x_test,y_train,y_test=train_test_split(onehot,y,test_size=0.3,random_state=42)
```

```python
model2=DecisionTreeClassifier()
```

```python
model2.fit(x_train,y_train)
```

```python
y_pred=model2.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```
