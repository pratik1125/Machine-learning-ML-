```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
```

```python
data= pd.read_csv("pets.csv")
data
```

```python
data.info()
```

```python
x= data.drop(['Can we pet them'],axis=1)
x
```

```python
y= data['Can we pet them']
y
```

```python
binary= pd.get_dummies(x)
binary
```

```python
model= BernoulliNB()
model.fit(binary,y)
```

```python
model.predict([[0,0,1,1,0,0,0,0,1]])
```
