```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import BernoulliNB
```

```python
df=pd.read_csv("flu .csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
x=df.drop(['Flue'],axis=1)#input colum
```

```python
x
```

```python
y=df['Flue']#output column
y
```

```python
Bernb=BernoulliNB()
```

```python
#Bernb.fit(x,y)
```

```python
one_hot_data=pd.get_dummies(x)
```

```python
one_hot_data
```

```python
train_df=Bernb.fit(one_hot_data,y)
```

```python
predict=train_df.predict([[0,1,1,0,1,0,0,0,1]])
```
