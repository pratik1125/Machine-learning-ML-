```python
#Series Using Array
data=np.array([33,44,55,66])
s2=pd.Series(data)
s2
```

```python
#Datadrame with deafault index
apple=[10,20,30,40]
banana=[11,22,33,44]

df1=pd.DataFrame([apple,banana])
df1
```

```python
#Datadrame with labbled index
apple=[10,20,30,40]
banana=[11,22,33,44]

df2=pd.DataFrame([apple,banana],index=['apple','banana'],columns=['jan','feb','march','april'])
df2
```

```python
#Create dataframe using multiple list of different length
apple=[10,20,30,40]
banana=[11,22,33,44]
orange=[55,66,77,88,99]

rw=['apple','banana','orange']
cl=['jan','feb','march','april','may']
df2=pd.DataFrame([apple,banana,orange],index=rw,columns=cl)
df2
```
