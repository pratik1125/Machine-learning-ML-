```python
import numpy as np
import pandas as pd
!pip install apyori
from apyori import apriori
```

```python
df=pd.read_csv('/content/puchased_data - puchased_data.csv')
```

```python
df.head()
```

```python
df.isnull()
```

```python
len(df)
```

```python
records=[]
for i in range(0,len(df)):
  temp=[]
  for j in range(0,4): #0 is strating index and 4 is number of columns or ending index
    if str(df.values[i,j])!="nan":
      temp.append(str(df.values[i,j]))
      records.append(temp)
print(records)

```

```python
#Generate rules
association_rules=apriori(records,min_support=0.5,min_confidence=0.75)
association_results=list(association_rules)
print(association_results)
```
